```python
import nltk
import matplotlib.pyplot as plt
from nltk import FreqDist
from nltk import word_tokenize
from nltk.corpus import gutenberg
```

```python
nltk.download('gutenberg')
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
text = gutenberg.raw('austen-persuasion.txt')

tokens = word_tokenize(text)

target_words = ['love', 'marriage', 'family', 'society', 'friendship']
```

```python
plt.figure(figsize=(15, 10))
nltk.draw.dispersion_plot(tokens, target_words)

plt.title('Dispersion Plot of Selected Words in "Persuasion"')
plt.xlabel('Word position in text')
plt.ylabel('Words')
plt.grid(True)
plt.show()
```
