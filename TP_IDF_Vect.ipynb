```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
```

```python
documents = ["Machine learning is a subset of artificial intelligence.",
            "Systems learn and improve from data without being explicitly programmed",
            "Iinvolves using algorithms to analyze large datasets, identify patterns, and make decisions or predictions",
            "Common applications include recommendation engines, spam filters, and image recognition. "]
```

```python
vectorizer = TfidfVectorizer()
```

```python
tfidf_matrix = vectorizer.fit_transform(documents)
```

```python
tfidf_dense = tfidf_matrix.todense()
```

```python
feature_names = vectorizer.get_feature_names_out()
```

```python
tfidf_df = pd.DataFrame(tfidf_dense, columns=feature_names)
print("TF-IDF Matrix:\n", tfidf_df)
```
