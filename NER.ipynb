```python
import spacy
from spacy import displacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
text = """Apple is lookong at buying a UK startup for $1 million.
Elon Musk announced that Tesla will build a factory in Berlin next year."""
```

```python
doc = nlp(text)
```

```python
print("Named Enities and Labels:")
for ent in doc.ents:
  print(ent.text, ent.label_)
```

```python
displacy.render(doc, style='ent', jupyter=True)
```
